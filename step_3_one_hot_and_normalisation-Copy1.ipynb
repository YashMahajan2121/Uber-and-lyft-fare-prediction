```python
import pandas as pd 

import numpy as np

from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1,inplace=True)
```

```python
df.describe()
```

```python
dff=df[["distance",'avg_travel_time','temp','clouds','pressure','rain','humidity','wind']]
```

```python
dff
```

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

dff=scalar.fit_transform(dff)

dff = pd.DataFrame(dff, columns=[
  "distance",'avg_travel_time','temp','clouds','pressure','rain','humidity','wind'])
```

```python
dff
```

```python
df.drop(["distance",'avg_travel_time','temp','clouds','pressure','rain','humidity','wind'], axis=1, inplace=True)
```

```python
df = pd.concat([dff, df], axis=1, join='inner')
```

```python
df.describe()
```

```python
# Columns to apply outlier detection on
columns_to_check = ['distance', 'avg_travel_time', 'temp', 'clouds', 'pressure', 'rain', 'humidity', 'wind']

# Loop through each column
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(upper_bound)
    
    # Replace outliers with bounds
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
```

```python
df.describe()
```

# One Hot encoding

```python
df = pd.get_dummies(df, columns = ['surge_multiplier','cab_type','name','day','class_hour'])
```
