```python
import datetime
#convert 13digit time stamp to datetime format
df['date_time']= pd.to_datetime(df['time_stamp']/1000, unit='s')
df['date']= df['date_time'].dt.date
df['day'] = df.date_time.dt.dayofweek
df['hour'] = df.date_time.dt.hour
```

```python
for i in range(0,637976):
    if df['hour'][i]>=0 and df['hour'][i]<=3:
        df['class_hour'][i]=0
    if df['hour'][i]>=4 and df['hour'][i]<=6:
        df['class_hour'][i]=1
    if df['hour'][i]>=7 and df['hour'][i]<=9:
        df['class_hour'][i]=2   
    if df['hour'][i]>=10 and df['hour'][i]<=12:
        df['class_hour'][i]=3       
    if df['hour'][i]>=13 and df['hour'][i]<=15:
        df['class_hour'][i]=4    
    if df['hour'][i]>=16 and df['hour'][i]<=18:
        df['class_hour'][i]=5     
    if df['hour'][i]>=19 and df['hour'][i]<=21:
        df['class_hour'][i]=6    
    if df['hour'][i]>=22 and df['hour'][i]<=24:
        df['class_hour'][i]=7    
```
